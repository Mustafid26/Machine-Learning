```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. VISUALISASI PIE CHART

```python
### visualisasi takaran susu
bahan = ('VITAMIN A','VITAMIN D3','KALSIUM','FOSFOR')
info = (20,15,15,20)
plt.pie(
    info,
    labels = bahan,
)
plt.show()
```

```python
### visualisasi takaran susu prosentase
bahan = ('VITAMIN A','VITAMIN D3','KALSIUM','FOSFOR')
info = (20,15,15,20)
plt.pie(
    info,
    labels = bahan,
    autopct='%1.1f%%',
)
plt.show()
```

```python
### visualisasi takaran susu warna dan mendorong pie
bahan = ('VITAMIN A','VITAMIN D3','KALSIUM','FOSFOR')
info = (20,15,15,20)
color = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0,0,0.1,0)
plt.pie(
    info,
    labels = bahan,
    autopct ='%1.1f%%',
    explode = explode
)
plt.show()
```
